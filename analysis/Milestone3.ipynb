## Question 1: Does having a college education and above give you a higher chance of earning 50k a year?

####  Exploring this dataset will help understand whether it is generally a good idea for young adults to explore university as an option. As the cost of university rises year over year with increasing student debt, exploring this datset will help students make the right decision, income wise, for their future.

#### First, we must convert the income column into a numeric value so that we could get a percentage of individuals with different education qualifications that are making over 50k a year. By converting those below 50k income to 0 and those with 50k income to 1, we get a 0 to 1 ratio. 

#### At the bottom of the ratio (0), there are no one making over 50k. At the top of the ratio (1), everyone is making over 50k.

```python
income2 = (
        income
        .replace(' <=50K',0)
        .replace(' >50K',1)
)
income2
```

```python
graph1 = income2.groupby('education')['income'].mean().reset_index()
graph1=graph1.sort_values('income', ascending=False)
graph1=graph1.reset_index(drop= True)
graph1
```

```python
ax = sns.scatterplot(x='income', y="education",data=graph1)
ax.set(xlabel='Income Ratio (0-1)', ylabel='Education', title = 'Scatterplot Of Education Against Income Ratio')
```

#### Analysing The Graph
- As we can see based off the graph, there is a much higher chance of making over 50k income a year the more education one has.
- One particular point of focus lies in the difference between a high school graduate and a person with a Bachelors. 
- Someone with a bachelors are around 50% more likely to earn over 50k a year compared to someone that only has a high school graduate.
- The greatest chance of making over 50k are those with a doctorate at over 74%. 
- The lowest chance of making over 50k are those with a preschool education. Virtually no one with only a preschool education are making over 50k a year.

#### So is a college education a worthwhile endevaour? 

- Yes, as a college graduate with a Bachelors, an individual has a 50% higher chance of making over 50k a year in their lifetime compared to entering the workforce right away. 

## Question 2: Does being born a male or female impact your prospects of earning a 50k a year salary?

#### Understanding this will help prove whether the wage gap exist between males and females. As doing the same amount of work and getting lesser pay would be criminal in the modern world. 

#### First, we try to quantify the number of males and females in each occupation and income bracket by changing quantifying the sex column by changing Male to 0 and Female to 1. (0 - 100% Male, 1 - 100% Female)

```python
income3 = (
        income
        .replace(' Male',0)
        .replace(' Female',1)
)
income3
```

```python
graph2 = income3.groupby(['income'])['sex'].mean().reset_index()
graph2=graph2.sort_values('sex', ascending=False)
graph2=graph2.reset_index(drop= True)
graph2
```

#### Based off the values, we can see that there are a greater proportion of the male population making over 50k a year compared to the female population. Does this mean the wage gap exists? 

#### However, diving deeper into the data tells a different story. 

```python
graph2 = income3.groupby(['occupation'])['sex'].mean().reset_index()
graph2=graph2.sort_values('sex', ascending=False)
graph2=graph2.reset_index(drop= True)
graph2
```

```python
ax1 = sns.boxplot(x="sex",y='occupation',data=graph2)
ax1.set(xlabel='Male To Female Ratio (0-1)', ylabel='Occupation', title = 'Boxplot Of Occupation Against Male To Female Ratio')
```

##### An interesting observation is that virtually private home servants skew heavily towards the female population. 

```python
graph3 = income2.groupby(['occupation'])['income'].mean().reset_index()
graph3=graph3.sort_values('income', ascending=False)
graph3=graph3.reset_index(drop= True)
graph3
```

```python
ax2 = sns.scatterplot(x='income', y="occupation",data=graph3)
ax2.set(xlabel='Income Ratio (0-1)', ylabel='Occupation', title = 'Boxplot Of Occupation Against Income Ratio')
```

#### Almost all private home servants don't make over 50k. 

#### This could be due to lifestyle choices where the females decide to be take care of their child instead of a career, thereby sacrificing their chance of making over 50k a year. 

#### However, this does not necessary explain the income differences.

```python
types = sns.boxplot(x=income["hours-per-week"],y=income["sex"],data=income)
types.set(xlabel = 'Hours Worked Per Week', ylabel = 'Gender', title = 'Boxplot Of Gender Against Hours Worked Per Week')
```

##### As we can see in the above Graph, males on average  work longer hours than their female counterpart, they could command a larger income due to more work done. 

#### So does the wage gap exist?

- This simple analysis could not answer the question fully as more details are required such as the income ranges for each occupation and experience in a company. 
- However, our analysis has provided some answers as to why it exist and why it may look the way it does.
- Due to the different lifestyle choices chosen by males and females, there arises a wage gap.
- As females decide to become mothers and care for their young, they are unable to focus on their career and taking care of their child at the same time. This results in a lower ability to command a higher income
- Males, on the other hand, can command a higher income due to the sheer number of hours worked and in fields that offer higher pay. This could explain why males, on average, are able to earn more than their female counterparts.

## Question 3: Does working in different sectors (workclass) affect your chances?

#### Exploring this question can help answer whether a certain workclass environment may be the best place for one to earn over 50k a year. This could help individuals decide where they should work at in order to maximize their income level.

#### To explore this differnece, we are using the same income ratio as in Research Question 1.

```python
graph4 = income2.groupby(['workclass'])['income'].mean().reset_index()
graph4=graph4.sort_values('income', ascending=False)
graph4=graph4.reset_index(drop= True)
graph4
```

```python
ax3 = sns.violinplot(x='income', y="workclass",data=graph4)
ax3.set(xlabel='Chance Of Making Over 50k Income A Year (0-1)', ylabel='Workclass', title = 'Violinplot Of Workclass Against Income')
```

##### Based off the graph, it can be safe to say that being self employed would bring the greatest likelihood for an individual to make over 50k a year. Followed up by working in the government. The least likelihood workclass environment to be earning over 50k a year is by working in the private sector. 

#### So which route leads to the highest potential for an individual to make over 50k a year?

##### By ranking them together we get the self employment route being the best way possible. And working in the private sector the worst way possible. 

##### Self employment might lead to the highest likelihood due to the absence of a safety net as the individual is required to make all the decision himself. Plus it allows for an uncapped ceiling as the individual gets paid what he is worth as determined by the market.

##### While those working in the private sector, may enjoy a safety net. Their destiny is not in their hands as they are under the direction of the company who may not have their best interest in mind.
